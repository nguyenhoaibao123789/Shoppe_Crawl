```python
import pandas as pd
import glob
```

```python
all_files=glob.glob("Data/Product_property/*.csv")
list_df=[]
for file in all_files:
    print(file)
    df=pd.read_csv(file,encoding='utf_16')
    list_df.append(df)
    print(df)
```

```python
df = pd.concat(list_df, axis=0, ignore_index=True)
df
```
